```python
import matplotlib.pyplot as plt

# Function to read points from a file and plot them
def plot_points_with_lines(filenameOriginal, filenameConvexHull):
    # Lists to store x and y coordinates
    x_original = []
    y_original = []

    # Read the file
    with open(filenameOriginal, 'r') as file:
        for line in file:
            # Strip whitespace and split by comma
            x, y = map(float, line.strip().split(','))
            x_original.append(x)
            y_original.append(y)

    # Lists to store x and y coordinates
    x_convexHull = []
    y_convexHull = []

    # Read the file
    with open(filenameConvexHull, 'r') as file:
        for line in file:
            # Strip whitespace and split by comma
            x, y = map(float, line.strip().split(','))
            x_convexHull.append(x)
            y_convexHull.append(y)


    # Plotting
    plt.figure(figsize=(8, 6))
    
    # Plot points with lines connecting them
    plt.plot(x_convexHull, y_convexHull, color='blue', linestyle='-', marker='o', label='Line connecting points')
    
    # Plot original points with a different color
    plt.scatter(x_original, y_original, color='red', marker='o', label='Original points')
    
    plt.title('2D Plot of Points with Connecting Lines')
    plt.xlabel('X Coordinates')
    plt.ylabel('Y Coordinates')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
plot_points_with_lines('puntos_aleatorios.txt', 'output.txt')
```
