```python
import matplotlib.pyplot as plt

def plot_points_with_lines(filenameOriginal, filenameConvexHull):
    x_original = []
    y_original = []

    with open(filenameOriginal, 'r') as file:
        for line in file:
            x, y = map(float, line.strip().split(','))
            x_original.append(x)
            y_original.append(y)

    x_convexHull = []
    y_convexHull = []

    with open(filenameConvexHull, 'r') as file:
        for line in file:
            x, y = map(float, line.strip().split(','))
            x_convexHull.append(x)
            y_convexHull.append(y)

    plt.figure(figsize=(8, 6))
    
    plt.plot(x_convexHull, y_convexHull, color='blue', linestyle='-', marker='o', label='Puntos del cierre convexo')
    
    plt.scatter(x_original, y_original, color='red', marker='o', label='Puntos originales')
    
    plt.title('Graham Scan Convex Hull')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
plot_points_with_lines('puntos_aleatorios1.txt', 'output1.txt')
```
