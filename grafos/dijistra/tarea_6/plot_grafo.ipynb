```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def generateGraphPlot(graphFilename, source = None):
    G = nx.DiGraph()
    
    graphTxt = open(graphFilename, "r")

    rowNo = 0
    line = graphTxt.readline()
    while (line != ''):
        row = line.split("\n")[0].split(" ")

        for i, weight in enumerate(row):
            if weight != "0":

                u = i
                v = rowNo

                G.add_edge(u, v, weight = int(weight))


        line = graphTxt.readline()

        rowNo+= 1

    pos = nx.spring_layout(G)

    color_map = []

    if (source == None):
        for node in G:
            color_map.append('blue')
    else:

        for node in G:
            if node == source:
                color_map.append('red')
            else: 
                color_map.append('blue')
            

    nx.draw(G, pos, with_labels=True, node_color = color_map)

    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.show()
```

# Grafo 1: 6 nodos 8 aristas

```python
generateGraphPlot("grafo1.txt")
generateGraphPlot("grafo1dijkstra.txt", 1)
```
