# Dependencias

```python
import matplotlib.pyplot as plt
import networkx as nx
```

# Funciones

```python
def generateGraphPlot(graphFilename, source = None, target = None, title = "Grafo"):
    G = nx.DiGraph()
    
    graphTxt = open(graphFilename, "r")

    rowNo = 0
    line = graphTxt.readline()
    while (line != ''):
        row = line.split("\n")[0].split(" ")

        for i, weight in enumerate(row):
            if weight != "0":

                u = rowNo
                v = i

                G.add_edge(u, v, weight = float(weight))


        line = graphTxt.readline()

        rowNo+= 1

    pos = nx.spring_layout(G, k=5)

    color_map = []

    if (source == None):
        for node in G:
            color_map.append('blue')
    else:

        for node in G:
            if node == source:
                color_map.append('red')
            elif node == target: 
                color_map.append('blue')
            else: 
                color_map.append('green')
            
    labels = nx.get_edge_attributes(G, 'weight')
    labels = {k: f'{v:.2f}' for k, v in labels.items()}

    nx.draw(G, pos, with_labels=True, node_color=color_map)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    plt.title(title)
    plt.show()
```

# Grafo 1, 10 nodos, nodo 0 a 5, 50% probabilidad de aristas

```python
generateGraphPlot("graph1.txt", title="Grafo original")
generateGraphPlot("graph1D.txt", 0, 5, title="Dijkstra, peso total: 56.44")
generateGraphPlot("graph1A.txt", 0, 5, title="A*, peso total: 56.44")
```
